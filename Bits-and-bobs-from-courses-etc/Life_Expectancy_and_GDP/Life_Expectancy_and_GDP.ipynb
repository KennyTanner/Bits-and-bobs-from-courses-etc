```python
sns.histplot(data['Life expectancy at birth (years)'])
```

```python
sns.pairplot(data, hue="Country")
```

```python
for country in data.Country.unique():
    plt.clf()
    plt.scatter(data.GDP[data["Country"]==country], data["Life expectancy at birth (years)"][data["Country"]==country])
    plt.xlabel('GDP (USD)')
    plt.ylabel('Life Expectancy at Birth (years)')
    plt.title(str(country))
    plt.show()
```
